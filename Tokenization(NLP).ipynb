```python
var = "To improve your data analysis skills, consider the following approaches:Develop a Strong Foundation: Master fundamental skills such as statistics, data visualization, and programming languages like Python and R.1Opt for Educational Programs: Pursue a degree, course, or bootcamp in data analytics to gain structured learning experiences."
```

```python
var
```

```python
from nltk.tokenize import word_tokenize , sent_tokenize
```

```python
sent = sent_tokenize(var)
```

```python
sent
```

```python
for i in sent:
    print(i)
    print()
```

```python
word = word_tokenize(var)
```
